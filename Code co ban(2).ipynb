```python
a=open('sequence.fasta').read()
b=a.splitlines()
header=''
seq=''
DNA=''
G=0
C=0
for each in b:
    if each =='>':
        header = each
    elif seq=='':
        seq = each
    elif seq !='':
        seq = seq + each
for count in list(seq):
    if count =='G':
        G=G+1
    elif count =='C':
        C=C+1
for nu in list(seq):
    if nu =='A':
        nu='T'
        DNA=DNA + nu
    elif nu =='T':
        nu='A'
        DNA=DNA + nu
    elif nu =='G':
        nu='C'
        DNA=DNA + nu
    elif nu =='C':
        nu='G'
        DNA=DNA + nu
print('1',len(seq))
print('2',(G+C)/len(seq))
print('3',DNA)
        
    
```
